```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Styling
sns.set_palette("Set2")
plt.rcParams['figure.figsize'] = (14, 6)
plt.rcParams['font.size'] = 11

print("="*70)
print(" SPOTIFY PREMIUM ADOPTION - DETAILED ANALYSIS")
print("="*70)

# 1. LOAD CLEANED DATA
print("Loading Cleaned Data...")

df = pd.read_csv('data/cleaned/spotify_cleaned.csv')

print(f"âœ“ Loaded: {df.shape[0]:,} rows Ã— {df.shape[1]} columns")

# 2. OVERALL METRICS
print("\n" + "="*70)
print("KEY METRICS OVERVIEW")
print("="*70)

total_users = len(df)
premium_users = df['is_premium_user'].sum()
premium_willing = df['is_premium_willing'].sum()
free_users = total_users - premium_users
free_willing = df[df['is_premium_user'] == 0]['is_premium_willing'].sum()

print(f"""
Total Users: {total_users:,}
â”œâ”€ Premium Users: {premium_users:,} ({premium_users/total_users*100:.1f}%)
â””â”€ Free Users: {free_users:,} ({free_users/total_users*100:.1f}%)

Premium Willingness:
â”œâ”€ Overall: {premium_willing:,} ({premium_willing/total_users*100:.1f}%)
â””â”€ Among Free Users: {free_willing:,} ({free_willing/free_users*100:.1f}%)

ðŸ’° Conversion Potential: {free_willing:,} users ready to upgrade!
""")

# 3. DEMOGRAPHIC ANALYSIS
print("\n" + "="*70)
print("ðŸ‘¥ DEMOGRAPHIC ANALYSIS")
print("="*70)

# Age Analysis
print("\n1. PREMIUM WILLINGNESS BY AGE GROUP")
print("-" * 50)

age_analysis = df.groupby('age').agg({
    'is_premium_willing': ['count', 'sum', 'mean']
}).round(3)
age_analysis.columns = ['Total_Users', 'Willing_Count', 'Willingness_Rate']
age_analysis['Willingness_Rate'] = age_analysis['Willingness_Rate'] * 100
age_analysis = age_analysis.sort_values('Willingness_Rate', ascending=False)

print(age_analysis)

# Visualization
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

# Chart 1: Willingness by Age
age_willing = df.groupby('age')['is_premium_willing'].mean() * 100
age_willing.plot(kind='barh', ax=axes[0], color='#1DB954')
axes[0].set_title('Premium Willingness Rate by Age Group', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Willingness Rate (%)')
axes[0].set_ylabel('Age Group')
# Chart 2: User Distribution by Age
age_dist = df['age'].value_counts()
age_dist.plot(kind='bar', ax=axes[1], color='#1ED760')
axes[1].set_title('User Distribution by Age', fontsize=14, fontweight='bold')
axes[1].set_xlabel('Age Group')
axes[1].set_ylabel('Number of Users')
axes[1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.savefig('outputs/visualizations/premium_by_age.png', dpi=300, bbox_inches='tight')
print("\nâœ“ Saved: premium_by_age.png")

# Gender Analysis
print("\n2. PREMIUM WILLINGNESS BY GENDER")
print("-" * 50)

gender_analysis = df.groupby('gender').agg({
    'is_premium_willing': ['count', 'mean']
}).round(3)
gender_analysis.columns = ['Total_Users', 'Willingness_Rate']
gender_analysis['Willingness_Rate'] = gender_analysis['Willingness_Rate'] * 100

print(gender_analysis)

# 4. USAGE PATTERN ANALYSIS
print("\n" + "="*70)
print("USAGE PATTERN ANALYSIS")
print("="*70)

# Tenure Analysis
print("\n1. PREMIUM WILLINGNESS BY USER TENURE")
print("-" * 50)

tenure_analysis = df.groupby('tenure_category').agg({
    'is_premium_willing': ['count', 'mean']
}).round(3)
tenure_analysis.columns = ['Total_Users', 'Willingness_Rate']
tenure_analysis['Willingness_Rate'] = tenure_analysis['Willingness_Rate'] * 100
tenure_analysis = tenure_analysis.sort_values('Willingness_Rate', ascending=False)

print(tenure_analysis)

# Engagement Analysis
print("\n2. PREMIUM WILLINGNESS BY ENGAGEMENT LEVEL")
print("-" * 50)

freq_analysis = df.groupby('music_lis_frequency').agg({
    'is_premium_willing': ['count', 'mean']
}).round(3)
freq_analysis.columns = ['Total_Users', 'Willingness_Rate']
freq_analysis['Willingness_Rate'] = freq_analysis['Willingness_Rate'] * 100
freq_analysis = freq_analysis.sort_values('Willingness_Rate', ascending=False)

print(freq_analysis)

# Device Analysis
print("\n3. PREMIUM WILLINGNESS BY DEVICE")
print("-" * 50)

device_analysis = df.groupby('spotify_listening_device').agg({
    'is_premium_willing': ['count', 'mean']
}).round(3)
device_analysis.columns = ['Total_Users', 'Willingness_Rate']
device_analysis['Willingness_Rate'] = device_analysis['Willingness_Rate'] * 100
device_analysis = device_analysis.sort_values('Willingness_Rate', ascending=False)

print(device_analysis)
```

```python
# 5. CONTENT PREFERENCE ANALYSIS
print("\n" + "="*70)
print("CONTENT PREFERENCE ANALYSIS")
print("="*70)

# Music vs Podcast vs Both
print("\n1. CONTENT TYPE PREFERENCE")
print("-" * 50)

content_analysis = df.groupby('content_diversity').agg({
    'is_premium_willing': ['count', 'mean']
}).round(3)
content_analysis.columns = ['Total_Users', 'Willingness_Rate']
content_analysis['Willingness_Rate'] = content_analysis['Willingness_Rate'] * 100
content_analysis = content_analysis.sort_values('Willingness_Rate', ascending=False)

print(content_analysis)

# Podcast Impact
print("\n2. PODCAST ENGAGEMENT IMPACT")
print("-" * 50)

podcast_analysis = df.groupby('is_podcast_listener').agg({
    'is_premium_willing': ['count', 'mean']
}).round(3)
podcast_analysis.columns = ['Total_Users', 'Willingness_Rate']
podcast_analysis['Willingness_Rate'] = podcast_analysis['Willingness_Rate'] * 100
podcast_analysis.index = ['No Podcast Listening', 'Active Podcast Listener']

print(podcast_analysis)

# Visualization
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Chart 1: Content Type
content_willing = df.groupby('content_diversity')['is_premium_willing'].mean() * 100
content_willing.plot(kind='bar', ax=axes[0], color=['#1DB954', '#1ED760', '#FF6B6B'])
axes[0].set_title('Willingness by Content Type', fontsize=14, fontweight='bold')
axes[0].set_ylabel('Willingness Rate (%)')
axes[0].tick_params(axis='x', rotation=45)

# Chart 2: Top Music Genres
top_genres = df['fav_music_genre'].value_counts().head(8)
top_genres.plot(kind='barh', ax=axes[1], color='#1DB954')
axes[1].set_title('Top 8 Music Genres', fontsize=14, fontweight='bold')
axes[1].set_xlabel('Number of Users')

# Chart 3: Podcast Frequency Distribution
pod_freq = df['pod_lis_frequency'].value_counts()
axes[2].pie(pod_freq.values, labels=pod_freq.index, autopct='%1.1f%%', startangle=90)
axes[2].set_title('Podcast Listening Frequency', fontsize=14, fontweight='bold')

plt.tight_layout()
plt.savefig('outputs/visualizations/content_preferences.png', dpi=300, bbox_inches='tight')
print("\nâœ“ Saved: content_preferences.png")
# 6. SATISFACTION ANALYSIS
print("\n" + "="*70)
print("â­ SATISFACTION & RECOMMENDATION ANALYSIS")
print("="*70)

# Recommendation Rating Impact
print("\n1. PREMIUM WILLINGNESS BY RECOMMENDATION RATING")
print("-" * 50)

rating_analysis = df.groupby('music_recc_rating').agg({
    'is_premium_willing': ['count', 'mean']
}).round(3)
rating_analysis.columns = ['Total_Users', 'Willingness_Rate']
rating_analysis['Willingness_Rate'] = rating_analysis['Willingness_Rate'] * 100

print(rating_analysis)

# High Satisfaction Impact
print("\n2. HIGH SATISFACTION (4+ RATING) IMPACT")
print("-" * 50)

satisfaction_impact = df.groupby('high_satisfaction').agg({
    'is_premium_willing': ['count', 'mean']
}).round(3)
satisfaction_impact.columns = ['Total_Users', 'Willingness_Rate']
satisfaction_impact['Willingness_Rate'] = satisfaction_impact['Willingness_Rate'] * 100
satisfaction_impact.index = ['Low Satisfaction (1-3)', 'High Satisfaction (4-5)']

print(satisfaction_impact)

# Visualization
fig, ax = plt.subplots(1, 1, figsize=(10, 6))

rating_willing = df.groupby('music_recc_rating')['is_premium_willing'].mean() * 100
rating_willing.plot(kind='line', marker='o', linewidth=3, markersize=10, ax=ax, color='#1DB954')
ax.set_title('Premium Willingness by Recommendation Rating', fontsize=16, fontweight='bold')
ax.set_xlabel('Recommendation Rating', fontsize=12)
ax.set_ylabel('Willingness Rate (%)', fontsize=12)
ax.grid(True, alpha=0.3)
ax.set_xticks(rating_willing.index)

plt.tight_layout()
plt.savefig('outputs/visualizations/satisfaction_impact.png', dpi=300, bbox_inches='tight')
print("\nâœ“ Saved: satisfaction_impact.png")

# 7. USER SEGMENTATION
print("\n" + "="*70)
print("ðŸ‘¤ USER SEGMENTATION ANALYSIS")
print("="*70)

# Create user segments
def segment_users(row):
    # Power Users: Daily listeners + high engagement
    if 'Daily' in str(row['music_lis_frequency']) and row['engagement_score'] >= 8:
        return 'Power Users'
    # Content Explorers: Listen to both music and podcasts
    elif row['content_diversity'] == 'Both':
        return 'Multi-Content Consumers'
    # Music Purists: Only music, no podcasts
    elif row['is_podcast_listener'] == 0:
        return 'Music-Only Users'
    # Casual Users: Low engagement
    elif row['engagement_score'] <= 3:
        return 'Casual Users'
    # At-Risk: Low satisfaction
    elif row['high_satisfaction'] == 0:
        return 'At-Risk Users'
    else:
        return 'Standard Users'

df['user_segment'] = df.apply(segment_users, axis=1)

segment_analysis = df.groupby('user_segment').agg({
    'is_premium_willing': ['count', 'mean'],
    'is_premium_user': 'mean'
}).round(3)
segment_analysis.columns = ['Total_Users', 'Premium_Willingness_Rate', 'Current_Premium_Rate']
segment_analysis['Premium_Willingness_Rate'] = segment_analysis['Premium_Willingness_Rate'] * 100
segment_analysis['Current_Premium_Rate'] = segment_analysis['Current_Premium_Rate'] * 100
segment_analysis = segment_analysis.sort_values('Premium_Willingness_Rate', ascending=False)

print("\nUSER SEGMENTS:")
print(segment_analysis)

# Visualization
fig, ax = plt.subplots(1, 1, figsize=(12, 6))

segment_data = df.groupby('user_segment')['is_premium_willing'].mean() * 100
segment_data = segment_data.sort_values(ascending=True)
segment_data.plot(kind='barh', ax=ax, color='#1DB954')
ax.set_title('Premium Willingness by User Segment', fontsize=16, fontweight='bold')
ax.set_xlabel('Willingness Rate (%)', fontsize=12)
ax.set_ylabel('User Segment', fontsize=12)

for i, v in enumerate(segment_data.values):
    ax.text(v + 1, i, f'{v:.1f}%', va='center', fontweight='bold')

plt.tight_layout()
plt.savefig('outputs/visualizations/user_segments.png', dpi=300, bbox_inches='tight')
print("\nâœ“ Saved: user_segments.png")

# 8. SAVE UPDATED DATASET WITH SEGMENTS

print("\n" + "="*70)
print("ðŸ’¾ SAVING UPDATED DATASET")
print("="*70)

# Save with new segment column
df.to_csv('data/cleaned/spotify_cleaned.csv', index=False)
print("âœ“ Saved updated dataset with user_segment column")

# 9. KEY INSIGHTS SUMMARY
print("\n" + "="*70)
print("ðŸ’¡ KEY INSIGHTS SUMMARY")
print("="*70)

# Calculate key metrics
top_age = age_analysis.idxmax()['Willingness_Rate']
top_content = content_analysis.idxmax()['Willingness_Rate']
top_device = device_analysis.idxmax()['Willingness_Rate']
top_segment = segment_analysis.idxmax()['Premium_Willingness_Rate']

avg_willingness = df['is_premium_willing'].mean() * 100
high_sat_boost = satisfaction_impact.loc['High Satisfaction (4-5)', 'Willingness_Rate'] - \
                 satisfaction_impact.loc['Low Satisfaction (1-3)', 'Willingness_Rate']

print(f"""
ðŸŽ¯ TOP PERFORMING SEGMENTS:
  â€¢ Age Group: {top_age} ({age_analysis.loc[top_age, 'Willingness_Rate']:.1f}% willing)
  â€¢ Content Type: {top_content} ({content_analysis.loc[top_content, 'Willingness_Rate']:.1f}% willing)
  â€¢ Device: {top_device} ({device_analysis.loc[top_device, 'Willingness_Rate']:.1f}% willing)
  â€¢ User Segment: {top_segment} ({segment_analysis.loc[top_segment, 'Premium_Willingness_Rate']:.1f}% willing)

ðŸ“Š KEY DRIVERS:
  â€¢ High Satisfaction (4+ Rating) increases willingness by {high_sat_boost:.1f} percentage points
  â€¢ Multi-content consumers show highest premium intent
  â€¢ Engagement frequency strongly correlates with willingness

ðŸ’° BUSINESS OPPORTUNITY:
  â€¢ {free_willing:,} free users ready to convert ({free_willing/free_users*100:.1f}% of free base)
  â€¢ Focus on {top_segment} segment for highest ROI
  â€¢ Improve recommendations to boost satisfaction â†’ conversions
""")
# 10. GENERATE SUMMARY REPORT
print("\n" + "="*70)
print("ðŸ“„ GENERATING SUMMARY REPORT")
print("="*70)

# Create summary report
summary_report = f"""
# SPOTIFY PREMIUM ADOPTION ANALYSIS - SUMMARY REPORT
Generated: {pd.Timestamp.now().strftime('%Y-%m-%d %H:%M:%S')}

## DATASET OVERVIEW
- Total Users: {total_users:,}
- Premium Users: {premium_users:,} ({premium_users/total_users*100:.1f}%)
- Free Users: {free_users:,} ({free_users/total_users*100:.1f}%)
- Premium Willingness: {premium_willing:,} ({avg_willingness:.1f}%)

## TOP INSIGHTS

### 1. Demographics
- Best Age Group: {top_age} with {age_analysis.loc[top_age, 'Willingness_Rate']:.1f}% willingness
- Gender Distribution: {', '.join([f"{g}: {c}" for g, c in df['gender'].value_counts().items()])}

### 2. Content Preferences
- Top Content Type: {top_content}
- Most Popular Genre: {df['fav_music_genre'].value_counts().index[0]}
- Podcast Listeners: {df['is_podcast_listener'].sum():,} ({df['is_podcast_listener'].mean()*100:.1f}%)

### 3. User Segments
- Highest Converting: {top_segment} ({segment_analysis.loc[top_segment, 'Premium_Willingness_Rate']:.1f}%)
- Largest Segment: {segment_analysis['Total_Users'].idxmax()} ({segment_analysis['Total_Users'].max():,} users)

### 4. Key Drivers
- Satisfaction Impact: +{high_sat_boost:.1f}pp for 4+ ratings
- Engagement Impact: Daily listeners show {freq_analysis.loc[freq_analysis.index[0], 'Willingness_Rate']:.1f}% willingness

## RECOMMENDATIONS
1. Target {top_segment} segment with personalized offers
2. Improve recommendation engine to increase 4+ ratings
3. Expand content in top-performing genres
4. Develop device-specific features for {top_device} users
5. Create conversion campaigns for high-engagement free users

## CONVERSION POTENTIAL
- Immediate Opportunity: {free_willing:,} users
- Estimated Revenue Impact: ${free_willing * 9.99:.2f}/month (assuming $9.99/user)
- Annual Revenue Potential: ${free_willing * 9.99 * 12:.2f}
"""

# Save report
with open('outputs/analysis_summary_report.txt', 'w') as f:
    f.write(summary_report)

print("âœ“ Summary report saved: outputs/analysis_summary_report.txt")

print("\n" + "="*70)
print("âœ… ANALYSIS COMPLETE!")
print("="*70)
print("\nGenerated Files:")
print("  âœ“ outputs/visualizations/premium_by_age.png")
print("  âœ“ outputs/visualizations/content_preferences.png")
print("  âœ“ outputs/visualizations/satisfaction_impact.png")
print("  âœ“ outputs/visualizations/user_segments.png")
print("  âœ“ outputs/analysis_summary_report.txt")
print("  âœ“ data/cleaned/spotify_cleaned.csv (updated)")
print("\nNext Steps:")
print("  1. Run notebook 04_export_to_sql.ipynb")
print("  2. Execute SQL queries in PostgreSQL")
print("  3. Create Power BI dashboard")
print("="*70)
```

```python
# 5. CONTENT PREFERENCE ANALYSIS
print("\n" + "="*70)
print("CONTENT PREFERENCE ANALYSIS")
print("="*70)

# Music vs Podcast vs Both
print("\n1. CONTENT TYPE PREFERENCE")
print("-" * 50)

content_analysis = df.groupby('content_diversity').agg({
    'is_premium_willing': ['count', 'mean']
}).round(3)
content_analysis.columns = ['Total_Users', 'Willingness_Rate']
content_analysis['Willingness_Rate'] = content_analysis['Willingness_Rate'] * 100
content_analysis = content_analysis.sort_values('Willingness_Rate', ascending=False)

print(content_analysis)

# Podcast Impact
print("\n2. PODCAST ENGAGEMENT IMPACT")
print("-" * 50)

podcast_analysis = df.groupby('is_podcast_listener').agg({
    'is_premium_willing': ['count', 'mean']
}).round(3)
podcast_analysis.columns = ['Total_Users', 'Willingness_Rate']
podcast_analysis['Willingness_Rate'] = podcast_analysis['Willingness_Rate'] * 100
podcast_analysis.index = ['No Podcast Listening', 'Active Podcast Listener']

print(podcast_analysis)

# Visualization
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Chart 1: Content Type
content_willing = df.groupby('content_diversity')['is_premium_willing'].mean() * 100
content_willing.plot(kind='bar', ax=axes[0], color=['#1DB954', '#1ED760', '#FF6B6B'])
axes[0].set_title('Willingness by Content Type', fontsize=14, fontweight='bold')
axes[0].set_ylabel('Willingness Rate (%)')
axes[0].tick_params(axis='x', rotation=45)

# Chart 2: Top Music Genres
top_genres = df['fav_music_genre'].value_counts().head(8)
top_genres.plot(kind='barh', ax=axes[1], color='#1DB954')
axes[1].set_title('Top 8 Music Genres', fontsize=14, fontweight='bold')
axes[1].set_xlabel('Number of Users')

# Chart 3: Podcast Frequency Distribution
pod_freq = df['pod_lis_frequency'].value_counts()
axes[2].pie(pod_freq.values, labels=pod_freq.index, autopct='%1.1f%%', startangle=90)
axes[2].set_title('Podcast Listening Frequency', fontsize=14, fontweight='bold')

plt.tight_layout()
plt.savefig('outputs/visualizations/content_preferences.png', dpi=300, bbox_inches='tight')
print("\nâœ“ Saved: content_preferences.png")
# 6. SATISFACTION ANALYSIS
print("\n" + "="*70)
print("â­ SATISFACTION & RECOMMENDATION ANALYSIS")
print("="*70)

# Recommendation Rating Impact
print("\n1. PREMIUM WILLINGNESS BY RECOMMENDATION RATING")
print("-" * 50)

rating_analysis = df.groupby('music_recc_rating').agg({
    'is_premium_willing': ['count', 'mean']
}).round(3)
rating_analysis.columns = ['Total_Users', 'Willingness_Rate']
rating_analysis['Willingness_Rate'] = rating_analysis['Willingness_Rate'] * 100

print(rating_analysis)

# High Satisfaction Impact
print("\n2. HIGH SATISFACTION (4+ RATING) IMPACT")
print("-" * 50)

satisfaction_impact = df.groupby('high_satisfaction').agg({
    'is_premium_willing': ['count', 'mean']
}).round(3)
satisfaction_impact.columns = ['Total_Users', 'Willingness_Rate']
satisfaction_impact['Willingness_Rate'] = satisfaction_impact['Willingness_Rate'] * 100
satisfaction_impact.index = ['Low Satisfaction (1-3)', 'High Satisfaction (4-5)']

print(satisfaction_impact)

# Visualization
fig, ax = plt.subplots(1, 1, figsize=(10, 6))

rating_willing = df.groupby('music_recc_rating')['is_premium_willing'].mean() * 100
rating_willing.plot(kind='line', marker='o', linewidth=3, markersize=10, ax=ax, color='#1DB954')
ax.set_title('Premium Willingness by Recommendation Rating', fontsize=16, fontweight='bold')
ax.set_xlabel('Recommendation Rating', fontsize=12)
ax.set_ylabel('Willingness Rate (%)', fontsize=12)
ax.grid(True, alpha=0.3)
ax.set_xticks(rating_willing.index)

plt.tight_layout()
plt.savefig('outputs/visualizations/satisfaction_impact.png', dpi=300, bbox_inches='tight')
print("\nâœ“ Saved: satisfaction_impact.png")

# 7. USER SEGMENTATION
print("\n" + "="*70)
print("ðŸ‘¤ USER SEGMENTATION ANALYSIS")
print("="*70)

# Create user segments
def segment_users(row):
    # Power Users: Daily listeners + high engagement
    if 'Daily' in str(row['music_lis_frequency']) and row['engagement_score'] >= 8:
        return 'Power Users'
    # Content Explorers: Listen to both music and podcasts
    elif row['content_diversity'] == 'Both':
        return 'Multi-Content Consumers'
    # Music Purists: Only music, no podcasts
    elif row['is_podcast_listener'] == 0:
        return 'Music-Only Users'
    # Casual Users: Low engagement
    elif row['engagement_score'] <= 3:
        return 'Casual Users'
    # At-Risk: Low satisfaction
    elif row['high_satisfaction'] == 0:
        return 'At-Risk Users'
    else:
        return 'Standard Users'

df['user_segment'] = df.apply(segment_users, axis=1)

segment_analysis = df.groupby('user_segment').agg({
    'is_premium_willing': ['count', 'mean'],
    'is_premium_user': 'mean'
}).round(3)
segment_analysis.columns = ['Total_Users', 'Premium_Willingness_Rate', 'Current_Premium_Rate']
segment_analysis['Premium_Willingness_Rate'] = segment_analysis['Premium_Willingness_Rate'] * 100
segment_analysis['Current_Premium_Rate'] = segment_analysis['Current_Premium_Rate'] * 100
segment_analysis = segment_analysis.sort_values('Premium_Willingness_Rate', ascending=False)

print("\nUSER SEGMENTS:")
print(segment_analysis)

# Visualization
fig, ax = plt.subplots(1, 1, figsize=(12, 6))

segment_data = df.groupby('user_segment')['is_premium_willing'].mean() * 100
segment_data = segment_data.sort_values(ascending=True)
segment_data.plot(kind='barh', ax=ax, color='#1DB954')
ax.set_title('Premium Willingness by User Segment', fontsize=16, fontweight='bold')
ax.set_xlabel('Willingness Rate (%)', fontsize=12)
ax.set_ylabel('User Segment', fontsize=12)

for i, v in enumerate(segment_data.values):
    ax.text(v + 1, i, f'{v:.1f}%', va='center', fontweight='bold')

plt.tight_layout()
plt.savefig('outputs/visualizations/user_segments.png', dpi=300, bbox_inches='tight')
print("\nâœ“ Saved: user_segments.png")

# 8. SAVE UPDATED DATASET WITH SEGMENTS

print("\n" + "="*70)
print("ðŸ’¾ SAVING UPDATED DATASET")
print("="*70)

# Save with new segment column
df.to_csv('data/cleaned/spotify_cleaned.csv', index=False)
print("âœ“ Saved updated dataset with user_segment column")

# -----------------------------
# 9. KEY INSIGHTS SUMMARY
# -----------------------------
print("\n" + "="*70)
print("ðŸ’¡ KEY INSIGHTS SUMMARY")
print("="*70)

# Find top segments
top_age = age_analysis['Willingness_Rate'].idxmax()
top_content = content_analysis['Willingness_Rate'].idxmax()
top_device = device_analysis['Willingness_Rate'].idxmax()
top_segment = segment_analysis['Premium_Willingness_Rate'].idxmax()

# Calculate satisfaction boost
high_sat_boost = satisfaction_impact.loc['High Satisfaction (4-5)', 'Willingness_Rate'] - \
                 satisfaction_impact.loc['Low Satisfaction (1-3)', 'Willingness_Rate']

avg_willingness = df['is_premium_willing'].mean() * 100

print(f"""
ðŸŽ¯ TOP PERFORMING SEGMENTS:
  â€¢ Age Group: {top_age} ({age_analysis.loc[top_age, 'Willingness_Rate']:.1f}% willing)
  â€¢ Content Type: {top_content} ({content_analysis.loc[top_content, 'Willingness_Rate']:.1f}% willing)
  â€¢ Device: {top_device} ({device_analysis.loc[top_device, 'Willingness_Rate']:.1f}% willing)
  â€¢ User Segment: {top_segment} ({segment_analysis.loc[top_segment, 'Premium_Willingness_Rate']:.1f}% willing)

ðŸ“Š KEY DRIVERS:
  â€¢ High Satisfaction (4+ Rating) increases willingness by {high_sat_boost:.1f} percentage points
  â€¢ Multi-content consumers show highest premium intent
  â€¢ Engagement frequency strongly correlates with willingness

ðŸ’° BUSINESS OPPORTUNITY:
  â€¢ {free_willing:,} free users ready to convert ({free_willing/free_users*100:.1f}% of free base)
  â€¢ Focus on {top_segment} segment for highest ROI
  â€¢ Improve recommendations to boost satisfaction â†’ conversions
""")

# -----------------------------
# 10. GENERATE SUMMARY REPORT
# -----------------------------
print("\n" + "="*70)
print("ðŸ“„ GENERATING SUMMARY REPORT")
print("="*70)

# Create summary report
summary_report = f"""
# SPOTIFY PREMIUM ADOPTION ANALYSIS - SUMMARY REPORT
Generated: {pd.Timestamp.now().strftime('%Y-%m-%d %H:%M:%S')}

## DATASET OVERVIEW
- Total Users: {total_users:,}
- Premium Users: {premium_users:,} ({premium_users/total_users*100:.1f}%)
- Free Users: {free_users:,} ({free_users/total_users*100:.1f}%)
- Premium Willingness: {premium_willing:,} ({avg_willingness:.1f}%)

## TOP INSIGHTS

### 1. Demographics
- Best Age Group: {top_age} with {age_analysis.loc[top_age, 'Willingness_Rate']:.1f}% willingness
- Gender Distribution: {', '.join([f"{g}: {c}" for g, c in df['gender'].value_counts().items()])}

### 2. Content Preferences
- Top Content Type: {top_content}
- Most Popular Genre: {df['fav_music_genre'].value_counts().index[0]}
- Podcast Listeners: {df['is_podcast_listener'].sum():,} ({df['is_podcast_listener'].mean()*100:.1f}%)

### 3. User Segments
- Highest Converting: {top_segment} ({segment_analysis.loc[top_segment, 'Premium_Willingness_Rate']:.1f}%)
- Largest Segment: {segment_analysis['Total_Users'].idxmax()} ({segment_analysis['Total_Users'].max():,} users)

### 4. Key Drivers
- Satisfaction Impact: +{high_sat_boost:.1f}pp for 4+ ratings
- Engagement Impact: Daily listeners show higher willingness

## RECOMMENDATIONS
1. Target {top_segment} segment with personalized offers
2. Improve recommendation engine to increase 4+ ratings
3. Expand content in top-performing genres
4. Develop device-specific features for {top_device} users
5. Create conversion campaigns for high-engagement free users

## CONVERSION POTENTIAL
- Immediate Opportunity: {free_willing:,} users
- Estimated Revenue Impact: ${free_willing * 9.99:.2f}/month (assuming $9.99/user)
- Annual Revenue Potential: ${free_willing * 9.99 * 12:,.2f}
"""

# Save report
with open('outputs/analysis_summary_report.txt', 'w', encoding='utf-8') as f:
    f.write(summary_report)

print("âœ“ Summary report saved: outputs/analysis_summary_report.txt")

print("\n" + "="*70)
print("âœ… ANALYSIS COMPLETE!")
print("="*70)
print("\nðŸ“ Generated Files:")
print("  âœ“ outputs/visualizations/premium_by_age.png")
print("  âœ“ outputs/visualizations/content_preferences.png")
print("  âœ“ outputs/visualizations/satisfaction_impact.png")
print("  âœ“ outputs/visualizations/user_segments.png")
print("  âœ“ outputs/analysis_summary_report.txt")
print("  âœ“ data/cleaned/spotify_cleaned.csv (updated with segments)")

print("\nðŸš€ Next Steps:")
print("  1. Run notebook 04_export_to_sql.ipynb")
print("  2. Execute SQL queries in PostgreSQL")
print("  3. Create Power BI dashboard")
print("="*70)
```
